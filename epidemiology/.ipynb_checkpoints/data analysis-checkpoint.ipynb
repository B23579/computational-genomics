# Abstract
Data collection occurred at visits 1 (baseline), 3, and 5.
The primary outcome of interest is gestational age at end of pregnancy. Additional outcomes include
birthweight, clinical measures of periodontal disease, and various microbiological and immunological
outcomes.

 Gestational age can be thought of as “time until
end of pregnancy,” for which certain survival analysis methods would be appropriate.
The study used a log-rank
test stratified by center to compare time until end of pregnancy for treatment and control groups

: for example, to
gestational age, dichotomized as “preterm” or “not preterm” according to a gestational age cutoff, or to
birthweight dichotomized as “low” or “high” at the 2500 g or other cutoff (2500 g would be in keeping with the
World Health Organization’s definition for low birth weight)

Preterm birth, defined as delivery before 37 weeks of gestation, is a growing problem. In some cases, preterm
birth can lead to infant death; in others, its consequences may include neurodevelopmental disabilities,
cognitive impairment, and/or respiratory disorders in the child

Many risk factors for preterm birth have already
been identified, including maternal age, drug use, and diabetes. However, such factors are exhibited in only
about half of preterm birth mothers, highlighting a need to expand our understanding of what contributes to
preterm birth risk.

Several observational studies have suggested an association between maternal periodontal disease and preterm
birth.

```python

import pandas as pd
from sklearn import preprocessing

x = tst.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
tst=pd.DataFrame(x_scaled,columns=tst.columns)
tst.head()
```

```python
control = tst[tst['treatment']==0.]
tretme = tst[tst['treatment']==1.]
print(len(control), len(tretme))
control.isna().sum()
```

```python
y_control= control["Birthweight"]
y_treatment = tretme["Birthweight"]
control = control.drop(["Birthweight"],axis=1)
tretme = tretme.drop(["Birthweight"],axis=1)
```

# T-Learner Model Training

```python
# Machine learning model
from lightgbm import LGBMRegressor
# Initiate the light GBM model

T_learner_c = LGBMRegressor()
T_learner_t = LGBMRegressor()
# Fit the model
T_learner_c.fit(control, y_control);
T_learner_t.fit(tretme, y_treatment);
```

# T-Learner Model Predictions

To make the treatment effect estimation, two separate predictions need to be made using the trained model:

In the first prediction, the treatment variable is set to 1 for all the samples. This gives us the predicted outcome values if all the samples received the treatment.
In the second prediction, the treatment variable is set to 0 for all the samples. This gives us the predicted outcome values if none of the samples received the treatment.


```python
# Set treatment value to 1
with_treatment = control.assign(treatment = 1)
# With treatment predictions
with_treatment_predict = T_learner_t.predict(with_treatment)
# Set treatment value to 0
without_treatment = tretme.assign(treatment = 0)
# With treatment predictions
without_treatment_predict = T_learner_c.predict(without_treatment)
```

# T-Learner Individual Treatment Effect (ITE) for control groupe

```python
# ITE
ite = y_control - with_treatment_predict
# Save ITE data in a pandas dataframe
ite_df = pd.DataFrame({'ITE': ite, 'with_treatment_predict':with_treatment_predict, 'control':y_control})
# Take a look at the data
ite_df
```

```python
# visualization
ite_df.hist(column ='ITE', bins=50, grid=True, figsize=(12, 8))
```

```python
# Calculate ATE
ATE = ite.mean()
# Print out results
print(f'The average treatment effect (ATE) is {ATE:.2f}')
```

```python
# Create the segmentation function
def segmentation(x):
    if x < -0.5:
        return 'sleeping-dog'
    elif x > 0.5:
        return 'persuadable'
    else:
        return 'no_impact'
# Apply the function 
ite_df['segment'] = ite_df['ITE'].apply(segmentation)
# Take a look at the data
ite_df
```

```python
# Visualization
sns.countplot(x=ite_df['segment'])
```

# T-Learner 

```python
tst=completed_dataset
x = tst.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
tst=pd.DataFrame(x_scaled,columns=tst.columns)

control = tst[tst['treatment']==0.]
tretme = tst[tst['treatment']==1.]
print(len(control), len(tretme))



y_control= control["Birthweight"]
y_treatment = tretme["Birthweight"]
control = control.drop(["Birthweight"],axis=1)
tretme = tretme.drop(["Birthweight"],axis=1)
```

```python
# training 
# Machine learning model
from lightgbm import LGBMRegressor
# Initiate the light GBM model

T_learner_c = LGBMRegressor()
T_learner_t = LGBMRegressor()
# Fit the model
T_learner_c.fit(control, y_control);
T_learner_t.fit(tretme, y_treatment);

# Model prediction
# Set treatment value to 1
with_treatment = control.assign(treatment = 1)
# With treatment predictions
with_treatment_predict = T_learner_t.predict(with_treatment)
# Set treatment value to 0
without_treatment = tretme.assign(treatment = 0)
# With treatment predictions
without_treatment_predict = T_learner_c.predict(without_treatment)

# individual treatment effect
# ITE
ite = y_control - with_treatment_predict
# Save ITE data in a pandas dataframe
ite_df = pd.DataFrame({'ITE': ite, 'with_treatment_predict':with_treatment_predict, 'control':y_control})
# visualization
ite_df.hist(column ='ITE', bins=50, grid=True, figsize=(12, 8))
```

```python
#Average treatment effect: 
# Calculate ATE
ATE = ite.mean()
# Print out results
print(f'The average treatment effect (ATE) is {ATE:.2f}')
```

```python
# Create the segmentation function
def segmentation(x):
    if x < -0.5:
        return 'sleeping-dog'
    elif x > 0.5:
        return 'persuadable'
    else:
        return 'no_impact'
# Apply the function 
ite_df['segment'] = ite_df['ITE'].apply(segmentation)
# Visualization
sns.countplot(x=ite_df['segment'])
```
